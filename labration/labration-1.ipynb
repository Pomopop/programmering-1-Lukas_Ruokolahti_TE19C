```python
import math 
import random 

innecirkel = 0 #varibel så man kan lägga till 1 varje gång en punkt är inne i cirkeln. Utan denna skulle for satsen itne fungera. 
hypotenusan = 0
simpunkter = 2000


for i in range (simpunkter): 
    y = random.uniform(-1,1) 
    x = random.uniform(-1,1)
    hypotenusan = math.sqrt((x**2) + (y**2))
    if hypotenusan < 1: # Om ett är större än hypotenusan så lägger man till +1 till variabel "innecirkel"
        innecirkel += 1


resultat = innecirkel/simpunkter #Delar punkter inne i cirkeln på antalet simulerade punkter. 
print(resultat)
```

```python
import math 
import random 
import matplotlib.pyplot as plt

innecirkel = 0

for i in range (10000): 
    y = random.uniform(-1,1) 
    x = random.uniform(-1,1) 
    hypotenusan = math.sqrt(x**2 + y**2)
    if hypotenusan < 1:
        innecirkel += 1
resultat = innecirkel/10000

svar = resultat *4 #gångrar resultatet från förra uppgiften (0,785) med 4.
print(svar)
#svaret är ungefär pi.  

# crikeln är ungefär 0,785* så stor som en kvadrat med samma diameter, 
#pi är cirkelns omkrets

#cirkel /kvadrat = (r**2 *pi/2r *2r) = (r*r*pi/4 * r * r) (stryk r:n) = pi/4 = 0,785. Detta betyder att cirkeln är 0.785* stor som en kvadrat med samma diameter. 
#0,785 är sambandet mellan cirkeln och kvadraten. 
```

```python
import math 
import random 
import matplotlib.pyplot as plt

innecirkel = 0

for i in range (2000): 
    y = random.uniform(-1,1) 
    x = random.uniform(-1,1) 
    svar = math.sqrt(x**2 + y**2)
    if svar < 1:
        plt.plot(x,y,'*b') #Denna målar grafen och sätter ut punkter som är inne i cirkeln i färgen blå. 
        innecirkel += 1
    else: #Om resultatet på if satsen inte är uppfyllt så hoppar koden över det steget och hoppar till else:.
        plt.plot(x,y,'*g') #Samma som den andra men plotar dem som är utanför grafen i grönt. 
```
